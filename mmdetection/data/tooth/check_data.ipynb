```python
# read test.json coco annotation file
# and check there is a image corresponding to annotation
# find mismatch between image and annotation
```

```python
# read test.json
import json

with open('./annotations/test.json') as f:
    data = json.load(f)

# read images
import os

image_dir = './test'
image_files = os.listdir(image_dir)

# check image file exists
for image in data['images']:
    file_name = image['file_name']
    if file_name not in image_files:
        print('image file not found: ', file_name)

# check annotation file exists
for annotation in data['annotations']:
    image_id = annotation['image_id']
    image = [x for x in data['images'] if x['id'] == image_id]
    if len(image) == 0:
        print('annotation not found: ', image_id)

# check image and annotation match
for image in data['images']:
    image_id = image['id']
    annotations = [x for x in data['annotations'] if x['image_id'] == image_id]
    if len(annotations) == 0:
        print('annotation not found: ', image_id)
    if len(annotations) != 1:
        print('multiple annotations found: ', image_id)

for annotation in data['annotations']:
    image_id = annotation['image_id']
    image = [x for x in data['images'] if x['id'] == image_id]
    if len(image) == 0:
        print('image not found: ', image_id)
    if len(image) != 1:
        print('multiple images found: ', image_id)

# check category id
category_ids = [x['id'] for x in data['categories']]
for annotation in data['annotations']:
    category_id = annotation['category_id']
    if category_id not in category_ids:
        print('category not found: ', category_id)

# check segmentation
for annotation in data['annotations']:
    segmentation = annotation['segmentation']
    if len(segmentation) == 0:
        print('segmentation not found: ', annotation['id'])

# check bbox
for annotation in data['annotations']:
    bbox = annotation['bbox']
    if len(bbox) == 0:
        print('bbox not found: ', annotation['id'])

# check area
for annotation in data['annotations']:
    area = annotation['area']
    if area == 0:
        print('area is 0: ', annotation['id'])

# check iscrowd
for annotation in data['annotations']:
    iscrowd = annotation['iscrowd']
    if iscrowd not in [0, 1]:
        print('iscrowd is not 0 or 1: ', annotation['id'])
        
# check id
annotation_ids = [x['id'] for x in data['annotations']]
for annotation in data['annotations']:
    id = annotation['id']
    if id not in annotation_ids:
        print('id not found: ', id)
        
# check image id
image_ids = [x['id'] for x in data['images']]
for image in data['images']:
    id = image['id']
    if id not in image_ids:
        print('id not found: ', id)
        
# check file_name
for image in data['images']:
    file_name = image['file_name']
    if file_name == '':
        print('file_name is empty: ', image['id'])
        
# check height
for image in data['images']:
    height = image['height']
    if height == 0:
        print('height is 0: ', image['id'])
        
# check width
for image in data['images']:
    width = image['width']
    if width == 0:
        print('width is 0: ', image['id'])
```
